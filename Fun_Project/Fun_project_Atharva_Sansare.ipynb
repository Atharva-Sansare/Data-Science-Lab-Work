```python
import random

def play_rps_full():
    choices = ["rock", "paper", "scissors"]
    user_score = 0
    computer_score = 0
    ties = 0
    round_num = 0

    print("=" * 40)
    print("   Welcome to Rock-Paper-Scissors!")
    print("=" * 40)

    while True:
        round_num += 1
        print(f"\n--- Round {round_num} ---")
        print(f"Score → You: {user_score}  |  Computer: {computer_score}  |  Ties: {ties}")

        user = input("Your choice (rock/paper/scissors) or 'quit' to stop: ").lower().strip()

        if user == "quit":
            break

        if user not in choices:
            print("⚠ Invalid input! Try again.")
            round_num -= 1
            continue

        computer = random.choice(choices)
        print(f"\nYou chose:      {user}")
        print(f"Computer chose: {computer}")

        if user == computer:
            print(">>> It's a tie!")
            ties += 1
        elif (user == "rock"     and computer == "scissors") or \
             (user == "scissors" and computer == "paper")   or \
             (user == "paper"    and computer == "rock"):
            print(">>> You win this round!")
            user_score += 1
        else:
            print(">>> Computer wins this round!")
            computer_score += 1

    # Final results
    print("\n" + "=" * 40)
    print("         FINAL RESULTS")
    print("=" * 40)
    print(f"Rounds played : {round_num - 1}")
    print(f"Your wins     : {user_score}")
    print(f"Computer wins : {computer_score}")
    print(f"Ties          : {ties}")

    if user_score > computer_score:
        print("\nOverall Winner: YOU! Congratulations!")
    elif computer_score > user_score:
        print("\nOverall Winner: Computer. Better luck next time!")
    else:
        print("\nOverall Result: It's a tie overall!")

play_rps_full()
```
