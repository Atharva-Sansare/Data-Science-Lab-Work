**1. Data Access and File Comparison**  
  
The dataset was loaded from three different file types: Excel (.xlsx),
CSV (.csv), and TXT (.txt). The Excel file loaded correctly with shape
(715, 18). However, the CSV and TXT files required additional parsing
adjustments due to formatting inconsistencies. The CSV file was not
properly comma-separated and behaved like a fixed-width file. The TXT
file also required separator and encoding adjustments.  
  
This demonstrates that file extensions do not always guarantee
consistent structure. While the actual data content remains the same,
different file formats may require preprocessing before analysis.
Therefore, the data source format can affect how easily the dataset is
loaded, but not the underlying information once correctly parsed.

**4. Data Visualization and Interpretation**  
  
Data visualization techniques were used to examine the distribution and
behavior of numerical variables within the dataset. Histograms, bar
charts, boxplots, and scatter plots were generated to better understand
central tendencies, variability, skewness, and relationships between
variables.  
  
Histograms (Age, Photos, Friends)  
  
Histograms were created for key numerical variables such as Age, Photos,
and Friends.  
  
The age histogram shows that the majority of users fall between 18 and
25 years old, indicating that the dataset is dominated by young
adults.  
  
The histogram for photo uploads reveals a positively skewed
distribution. While most users upload a moderate number of photos, a
small subset of users exhibit extremely high activity levels, as
reflected by the long right tail of the distribution.  
  
Similarly, the friend count histogram demonstrates right skewness. Most
users maintain moderate social networks, but a small number have
exceptionally large friend counts.  
  
These visual patterns confirm what was observed in the descriptive
statistics, particularly the large differences between mean and maximum
values.  
  
Frequency Plot (Gender Distribution)  
  
A bar chart was used to visualize the frequency distribution of the
Gender variable.  
  
The frequency plot shows a slightly higher representation of one
category (coded as 1), which aligns with the mean value observed in the
descriptive statistics. This visualization provides a clearer
understanding of categorical distribution compared to numerical
summaries alone.  
  
Boxplot (Photo Uploads)  
  
The boxplot highlights the presence of extreme outliers in the number of
photos uploaded.  
  
The large interquartile range and the extreme upper values confirm
significant variability in user engagement. This suggests that a small
number of highly active users drive much of the platform activity.  
  
Scatter Plot (Posts vs Friends)  
  
The scatter plot comparing Posts and Friends suggests a weak positive
relationship between posting activity and number of friends. However,
considerable dispersion indicates that high posting frequency does not
necessarily guarantee a larger social network.  
  
🔷 Importance of Data Visualization  
  
Data visualization enhances statistical analysis by:  
  
Revealing skewness in distributions  
  
Identifying outliers  
  
Showing variability and dispersion  
  
Highlighting relationships between variables  
  
Providing intuitive confirmation of numerical findings  
  
While descriptive statistics provide numerical summaries, visualizations
allow patterns to be interpreted more clearly and effectively.
